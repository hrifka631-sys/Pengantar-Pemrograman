```python
import numpy as np

def integrate_rec(f, a, b, n):
    # Implementation of the rectanlge method 
    h = (b-a)/n
    x = np.linspace(a, b, n+1) 
    i=0
    area=0
    while i<n:
        sup_rect =  f(x[i])*h
        area += sup_rect
        i += 1 
    return area
'''
We test the Rectangle method on the sine funcion were the definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_rec(np.sin, 0, np.pi/2, 5)
sup_10 = integrate_rec(np.sin, 0, np.pi/2, 10)
sup_100 = integrate_rec(np.sin, 0, np.pi/2, 100)

print('Using n=5, the rectangle method returns a value of {:.2f}'.format(sup_5))
print('Using n=10, the rectangle method returns a value of {:.2f}'.format(sup_10))
print('Using n=100, the rectangle method returns a value of {:.2f}'.format(sup_100))
     
'''
Output:
Using n=5, the rectangle method returns a value of 0.83
Using n=10, the rectangle method returns a value of 0.92
Using n=100, the rectangle method returns a value of 0.99
'''
 
```

```python
import numpy as np

def integrate_trap(f, a, b, n):
    # Implementation of the trapezoidal rule
    h = (b-a)/n
    x = np.linspace(a, b, n+1) 
    i=1
    area = h*(f(x[0]) + f(x[n]))/2
    while i<n:
        sup_rect = f(x[i])*h
        area += sup_rect 
        i += 1 
    return area

'''
We test the trapezoidal rule on the known sine funcion were the  
definite integral in the interval [0, %*\textcolor{codeviolet}{$\pi$}*)/2] is equal to 1. 
'''

sup_5 = integrate_trap(np.sin, 0, np.pi/2, 5)
sup_10 = integrate_trap(np.sin, 0, np.pi/2, 10)

print('Using n=5, the trapezoidal rule returns a value of {:.2f}'.format(sup_5))
print('Using n=10, the trapezoidal rule returns a value of {:.2f}'.format(sup_10))
   
'''
Output:
Using n=5, the trapezoidal rule returns a value of 0.99
Using n=10, the trapezoidal rule returns a value of 1.00
'''
 
```

```python
import numpy as np
from scipy import integrate

# 1. Definisikan Data
# x = [0, 4.5, 9] (3 titik, 2 interval/n=2)
x = np.linspace(0, 9, 3) 
y = x**2

# 2. Perhitungan Integral (Perbaikan Fungsi)
# 'trapz' sekarang menjadi 'trapezoid'
# 'simps' sekarang menjadi 'simpson'
sup_trapz = integrate.trapezoid(y, x)
sup_simps = integrate.simpson(y, x)

# 3. Menampilkan Output
print('Using n=2, the trapezoidal rule returns a value of {:.0f}'.format(sup_trapz))
print('Using n=2, the composite Simpson rule returns a value of {:.0f}'.format(sup_simps))

'''
Output:
Using n=2, the trapezoidal rule returns a value of 273
Using n=2, the composite Simpson rule returns a value of 243
'''
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapezoid # Versi terbaru menggantikan trapz

# 1. Membuat array jarak r
r = np.linspace(1, 6400, 6400)

def density():
    # Mendefinisikan lapisan-lapisan Bumi (Total harus 6400 titik)
    ro_inner_core = np.linspace(13100, 12800, 1220)
    ro_outer_core = np.linspace(12200, 9900, 2259)
    ro_lower_mantle = np.linspace(5600, 4400, 2171)
    ro_upper_mantle = np.linspace(4130, 3400, 720)
    ro_crust = np.linspace(3100, 2700, 30)
    
    # Menggabungkan semua lapisan menjadi satu array tunggal
    ro_final = np.concatenate((ro_inner_core, ro_outer_core, ro_lower_mantle, ro_upper_mantle, ro_crust))
    
    return ro_final

# 2. Memanggil fungsi density
ro = density()

# 3. Visualisasi
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(r, ro, label="Densities of Earth's Interior", color='darkred', linewidth=2)

# Menambahkan garis vertikal sebagai penanda batas lapisan (Opsional tapi membantu)
boundaries = [1220, 1220+2259, 1220+2259+2171, 1220+2259+2171+720]
for b in boundaries:
    ax.axvline(x=b, color='gray', linestyle='--', alpha=0.5)

ax.set_ylabel(r"Density [kg/m$^3$]")
ax.set_xlabel("Distance from Earth center r [km]")
ax.set_title("Model Densitas Interior Bumi (PREM-like)")
ax.legend()
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapezoid # Versi terbaru dari trapz

def gravity(r_km, ro):
    # Inisialisasi array g dengan ukuran yang sama dengan r
    g = np.zeros(len(r_km))
    G = 6.67408e-11  # Konstanta gravitasi
    
    # Konversi radius dari Km ke m untuk perhitungan fisika
    r_m = r_km * 1000 
    
    # Loop dimulai dari indeks 1 untuk menghindari pembagian dengan nol di pusat (r=0)
    for i in range(1, len(r_m)):
        # Ambil potongan data dari pusat hingga radius ke-i
        r_slice = r_m[:i+1]
        ro_slice = ro[:i+1]
        
        # Jarak saat ini (titik pengamatan)
        current_r = r_m[i]
    
        # Fungsi yang akan diintegralkan: rho(r) * r^2
        y = ro_slice * (r_slice**2)
        
        # Hitung integral menggunakan aturan trapesium
        y_int = trapezoid(y, r_slice)
        
        # Rumus gravitasi di dalam bola: g = (4 * pi * G / r^2) * integral(rho * r^2 dr)
        g[i] = ((4 * np.pi * G) / (current_r**2)) * y_int
           
    return g

# Pastikan array 'ro' sudah didefinisikan sebelumnya dari fungsi density() Anda
# r adalah array radius dari 1 ke 6400
g = gravity(r, ro)

# Visualisasi
fig, ax = plt.subplots(figsize=(8, 5))
ax.plot(r, g, color='blue', linewidth=2, label='Gravity Acceleration')
ax.grid(True, linestyle='--', alpha=0.6)
ax.set_ylabel(r'g [m/s$^2$]')
ax.set_xlabel('Distance from the Earth center r [km]')
ax.set_title('Percepatan Gravitasi di Dalam Interior Bumi')
ax.legend()

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import trapezoid # Pengganti trapz

def pressure(r_km, ro, g):
    # Inisialisasi array p dengan ukuran yang sama dengan r
    p = np.zeros(len(r_km))
    
    # Konversi radius dari Km ke m untuk perhitungan fisika (P = integral rho * g * dr)
    r_m = r_km * 1000
    
    # Loop untuk menghitung tekanan di setiap kedalaman
    # Tekanan pada radius r adalah berat kolom material di atasnya (dari r sampai permukaan R_earth)
    for i in range(0, len(r_m)):    
        r_slice = r_m[i:len(r_m)]   # Mengambil irisan dari radius r ke permukaan
        ro_slice = ro[i:len(r_m)]
        g_slice = g[i:len(r_m)]
        
        y = ro_slice * g_slice
        
        # Hitung integral. Karena r_slice membesar, hasilnya akan positif.
        p_val = trapezoid(y, r_slice)
        p[i] = p_val
        
    return p

# 1. Hitung tekanan dan konversi ke GPa (1 GPa = 1e9 Pascal)
p_gpa = pressure(r, ro, g) / 1e9 

# 2. Definisikan Kedalaman (Depth) dari permukaan (0 km di permukaan, 6400 km di pusat)
# r adalah radius dari pusat (1-6400), maka depth = 6400 - r
depth = 6400 - r

# 3. Visualisasi
fig, ax = plt.subplots(figsize=(8, 5))  
ax.plot(depth, p_gpa, color='darkgreen', linewidth=2, label='Pressure Profile')

ax.grid(True, linestyle='--', alpha=0.6)
ax.set_ylabel('Pressure [GPa]')
ax.set_xlabel('Depth from the Earth Surface [km]')
ax.set_title('Profil Tekanan Hidrostatis Interior Bumi')
ax.invert_xaxis() # Opsional: Membalik sumbu X agar 0 (permukaan) di kiri dan 6400 di kanan
ax.legend()

plt.tight_layout()
plt.show()
```
