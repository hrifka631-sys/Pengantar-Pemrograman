```python
import pandas as pd 
#Digunakan untuk mengelola data dalam bentuk tabel (DataFrame). Sangat kuat untuk membaca file Excel/CSV.
from scipy.stats.mstats import gmean, hmean 
# Mengambil fungsi khusus untuk menghitung Geometric Mean ($G = \sqrt[n]{x_1 \cdot x_2 \dots x_n}$) dan Harmonic Mean ($H = \frac{n}{\sum \frac{1}{x_i}}$).
import matplotlib.pyplot as plt 
#Library standar untuk membuat grafik dan visualisasi data di Python.

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces') 
#Perintah untuk membaca file Excel.
#memastikan Python mengambil data dari lembar kerja yang spesifik, bukan lembar pertama saja.

a_mean = my_dataset.Zr.mean() 
#Menghitung Arithmetic Mean (rata-rata biasa). Ini adalah jumlah semua nilai dibagi jumlah data. Paling umum digunakan tapi sensitif terhadap nilai ekstrem (outlier).
g_mean = gmean(my_dataset['Zr']) 
#Menghitung rata-rata geometrik. Dalam geokimia, ini sering dianggap lebih mewakili "nilai tengah" untuk elemen jejak yang distribusinya tidak normal (menceng).
h_mean = hmean(my_dataset['Zr']) 
#Menghitung rata-rata harmonik. Ini memberikan bobot lebih besar pada nilai yang kecil dan biasanya menghasilkan nilai terendah di antara ketiganya.

print ('-------')
print ('arithmetic mean')
print ("{0:.1f} [ppm]".format(a_mean))
print ('-------')

print ('geometric mean')
print ("{0:.1f} [ppm]".format(g_mean))
print ('-------')

print ('harmonic mean')
print ("{0:.1f} [ppm]".format(h_mean))
print ('-------')
#Menampilkan teks di konsol agar kita bisa melihat hasil angka perhitungannya.
#Mengatur format angka agar hanya menampilkan 1 angka di belakang koma (desimal) dan menambahkan satuan [ppm] (parts per million).

fig, ax = plt.subplots()
#Membuat "kanvas" dan "sumbu" (axis) untuk menggambar grafik.
ax.hist(my_dataset.Zr, bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.8) 
#ax.hist(...): Membuat Histogram.
#bins='auto': Menentukan jumlah batang grafik secara otomatis.
#density=True: Mengubah sumbu Y menjadi kepadatan probabilitas (bukan jumlah mentah), agar luas total area di bawah kurva adalah 1.
ax.axvline(a_mean, color='purple', label='Arithmetic mean', linewidth=2)
ax.axvline(g_mean, color='orange',  label='Geometric mean', linewidth=2)
ax.axvline(h_mean, color='green',  label='Harmonic mean', linewidth=2)
#ax.axvline(...): Singkatan dari Axis Vertical Line. Fungsinya menggambar garis tegak lurus dari atas ke bawah untuk menandai posisi ketiga rata-rata tersebut pada grafik.
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.legend()
#ax.legend(): Memunculkan kotak keterangan (label) sehingga kita tahu warna ungu itu garis apa, warna hijau garis apa, dan seterusnya.

'''
Output:
-------
arithmetic mean
365.4 [ppm]
-------
geometric mean
348.6 [ppm]
-------
harmonic mean
333.8 [ppm]
-------
'''
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Memuat data
# Pastikan file Excel berada di folder yang sama dengan script ini
try:
    my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')
except FileNotFoundError:
    print("Error: File 'Smith_glass_post_NYT_data.xlsx' tidak ditemukan.")
    exit()

# 2. Perhitungan Median
# Menggunakan dropna() untuk memastikan tidak ada nilai kosong yang mengganggu
data_zr = my_dataset['Zr'].dropna()
median_val = data_zr.median()

# 3. Menampilkan Output Teks
print('-------')
print('Median')
print("{0:.1f} [ppm]".format(median_val))
print('-------')

# 4. Visualisasi
fig, ax = plt.subplots(figsize=(8, 5))

# Histogram
# density=True mengubah sumbu Y menjadi probabilitas (luas total = 1)
ax.hist(data_zr, bins=20, density=True, edgecolor='k', label="Measurements Hist", alpha=0.7) 

# Garis vertikal untuk Median
# Menambahkan label nilai di legenda agar lebih informatif
ax.axvline(median_val, color='orange', label=f'Median ({median_val:.1f})', linewidth=2)

# Label Sumbu dan Estetika
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.set_title('Distribusi Konsentrasi Zr dengan Garis Median')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

variance =  my_dataset['Zr'].var()
stddev =  my_dataset['Zr'].std()

print ('-------')
print ('Variance')
print("{0:.0f} [square ppm]".format(variance))
print ('-------')
print ('Standard Deviation')
print("{0:.0f} [ppm]".format(stddev))
print ('-------')

fig, ax = plt.subplots()
ax.hist(my_dataset.Zr, bins= 20, density = True, edgecolor='k', label='Measurements Hist') 
ax.axvline(my_dataset['Zr'].mean()-stddev, color='purple', label=r'mean - 1$\sigma$', linewidth=2)
ax.axvline(my_dataset['Zr'].mean()+stddev, color='green', label=r'mean + 1$\sigma$', linewidth=2)
ax.axvspan(my_dataset['Zr'].mean()-stddev, my_dataset['Zr'].mean()+stddev, alpha=0.1, color='orange', label=r'mean $\pm$ 1$\sigma$')
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.legend()

'''
Output:
-------
Variance
14021 [square ppm]
-------
Standard Deviation
118 [ppm]
-------
'''
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1. Memuat data
my_dataset = pd.read_excel('Smith_glass_post_NYT_data.xlsx', sheet_name='Supp_traces')

# 2. Perhitungan Kuartil dan IQR
# Catatan: parameter 'interpolation' sekarang disarankan menggunakan 'method' pada Numpy versi terbaru
q1 = np.percentile(my_dataset['Zr'], 25, method='midpoint') 
q3 = np.percentile(my_dataset['Zr'], 75, method='midpoint') 

iqr = q3 - q1  # Interquartile range (IQR)

# 3. Menampilkan Output
print('-------')
print(f'Interquartile range (IQR): {iqr:.0f} [ppm]')
print('-------')

# 4. Visualisasi
fig, ax = plt.subplots(figsize=(10, 6))

# Histogram
ax.hist(my_dataset['Zr'], bins='auto', density=True, edgecolor='k', label='Measurements Hist', alpha=0.6) 

# Garis vertikal untuk Q1 dan Q3
ax.axvline(q1, color='purple', linestyle='--', label=f'Q1 ({q1:.1f})', linewidth=2)
ax.axvline(q3, color='green', linestyle='--', label=f'Q3 ({q3:.1f})', linewidth=2)

# Mengarsir area IQR (Interquartile Range)
ax.axvspan(q1, q3, alpha=0.2, color='orange', label=f'IQR Area ({iqr:.0f} ppm)')

# Label dan Judul
ax.set_xlabel('Zr [ppm]')
ax.set_ylabel('Probability density')
ax.set_title('Visualisasi Interquartile Range (IQR) pada Data Zr')
ax.legend()

plt.tight_layout()
plt.show()
```

```python
import numpy as np
from scipy.stats import skew

a_mean = my_dataset.Zr.mean()
median = my_dataset.Zr.median()
hist, bin_edges = np.histogram(my_dataset['Zr'], bins=20, density=True)
modal_value = (bin_edges[hist.argmax()] + bin_edges[hist.argmax()+1])/2
standard_deviation = my_dataset['Zr'].std()

a1 = (a_mean - modal_value) / standard_deviation 
a2 = 3 * (a_mean - median) / standard_deviation
g1 = skew(my_dataset['Zr'])

print ('-------')
print ("Pearson's first coefficient of skewness: {:.2f}".format(a1))
print ("Pearson's 2nd moment of skewness: {:.2f}".format(a2))
print ("Fisher-Pearson's coefficient of skewness: {:.2f}".format(g1))
print ('-------')

'''
Output:
-------
Pearson's first coefficient of skewness: 0.74
Pearson's 2nd moment of skewness: 0.66
Fisher-Pearson's coefficient of skewness: 1.26
-------
'''
```
